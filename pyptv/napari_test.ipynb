```python
import napari
import imageio
```

```python
# let's use the parameters to get which 4 points need to be selected and in what order
import os
import numpy as np
manual_points_per_camera = np.loadtxt('../tests/test_cavity/parameters/man_ori.par')
print(manual_points_per_camera)
```

```python
# plot the calibration target to understand the location of the grid points
# as typed in the calibration target file
calibration_target = np.loadtxt('../tests/test_cavity/cal/target_on_a_side.txt')

import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(6, 6))
ax.plot(calibration_target[:, 1], calibration_target[:, 2], 'o');
for row in calibration_target:
    # print(row)
    ax.text(row[1], row[2], str(row[0].astype(int)),fontsize=12);
    if row[0] in manual_points_per_camera[:4]:
        ax.plot(row[1], row[2], 'o', color='red', markersize=10);
```
